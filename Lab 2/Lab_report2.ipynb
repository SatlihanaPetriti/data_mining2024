4. Постройте транзакционную базу данных из полученного датафрейма, используя в качестве идентификатора транзакции столбец order_id, а в качестве названий товаров - поле product_name из датафрейма для таблицы products, соответствующее столбцу product_id. Найдите в транзакционной базе данных транзакцию с наибольшим количеством товаров и выведите ее на экран.


```python
# Соединение таблицу filtered_df с таблицей products для получения названий товаров (product_name)
transaction_df = pd.merge(filtered_df, prod_df[['product_id', 'product_name']], on='product_id')
# Группировка по order_id и собираем названия товаров для каждой транзакции
transaction_db = transaction_df.groupby('order_id')['product_name'].apply(list).reset_index()
# Добавление колонку с количеством товаров в каждой транзакции
transaction_db['num_products'] = transaction_db['product_name'].apply(len)
# Находим транзакцию с наибольшим количеством товаров
max_products_transaction = transaction_db.loc[transaction_db['num_products'].idxmax()]
print("Транзакция с наибольшим количеством товаров:")
print(max_products_transaction)
```

5. Постройте по транзакционной базе данных бинарную базу данных в формате датафрейма пакета mlxtend. По бинарной базе данных определите и выведите на экран три наиболее популярных товара и определите количество покупок (транзакций) этих товаров.


```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Создание списка транзакций для каждого order_id, содержащего список приобретенных товаров
transactions = transaction_df.groupby('order_id')['product_name'].apply(list).tolist()
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
# Создание бинарный датафрейм
binary_df = pd.DataFrame(te_ary, columns=te.columns_)
# Подсчитывание количество транзакций для каждого товара
item_counts = binary_df.sum().sort_values(ascending=False)
# Определение три самых популярных товара
top_3_items = item_counts.head(3)
print("Три самых популярных товара и количество транзакций для каждого:")
print(top_3_items)
```

6. При помощи указанного в индивидуальном задании метода построения популярных наборов предметов постройте и выведите на экран популярный набор предметов с (абсолютной) минимальной поддержкой не менее 3, имеющий максимальную длину. При отсутствии такого набора уменьшите поддержку до 2. В случае нехватки вычислительных ресурсов (слишком долгой работы программы) при построении популярного набора предметов сокращайте число записей в наборе данных (например, делая выборку половины записей набора).


```python
from mlxtend.frequent_patterns import fpgrowth, fpmax
from mlxtend.frequent_patterns import association_rules
```

```python
from mlxtend.frequent_patterns import fpmax
import pandas as pd

# устанавливание минимальную поддержку: 3 транзакции
min_support = 3 / len(binary_df)

# Применения алгоритм FPMax для нахождения частых наборов предметов с заданной поддержкой
frequent_itemsets = fpmax(binary_df, min_support=min_support, use_colnames=True)
# Если не найдено ни одного набора предметов с поддержкой 3, уменьшаем поддержку до 2 и повторяем поиск
if frequent_itemsets.empty:
    min_support = 2 / len(binary_df)
    frequent_itemsets = fpmax(binary_df, min_support=min_support, use_colnames=True)

# Нахождение максимальной длины наборов предметов:
max_length = frequent_itemsets['itemsets'].apply(len).max()

# Извлечение наборов предметов с максимальной длиной:
max_length_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == max_length]

print("Популярный набор предметов с минимальной поддержкой и максимальной длиной:")
print(max_length_itemsets)
```

7. Используя пакет mlxtend или реализацию на Python, постройте набор ассоциативных правил для полученного популярного наборов предметов с уровнем достоверности (confidence) не ниже .


```python
from mlxtend.frequent_patterns import apriori, association_rules
min_support = 0.005  # Минимальная поддержка
# Вычисление частые наборы предметов с помощью apriori
frequent_itemsets = apriori(binary_df, min_support=min_support, use_colnames=True)
# Печатание структуру частых наборов
print(frequent_itemsets.head())
# Генерирование ассоциативные правила с заданным уровнем достоверности
min_confidence = 0.5  # Минимальная достоверность
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
print("Ассоциативные правила с заданным уровнем достоверности:")
print(rules)
```

8. Для построенного набора ассоциативных правил вычислите показатель (меру) оценки ассоциативных правил, указанную в индивидуальном задании, и выведите на экран ассоциативное правило, имеющее наилучшее значение этого показателя оценки.


```python
best_r = rules.sort_values('lift', ascending=False).iloc[0]
print("Ассоциативное правило с наилучшим значением лифта:")
print(best_r)
```
